```python
#plotting a very simple line graph
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, np.pi*4, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
#adding labels, title and legend
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='cos(x)')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend(loc='lower right')
```

```python
#Plotting Bar Plot
```

```python
plt.style.use('ggplot')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [400,340,544,688,921]

plt.bar(months, sales, width=0.5, label="Sales")
plt.title('Car Sales')
plt.legend()
```

```python
#pie plot
```

```python
plt.style.use('fivethirtyeight')
revenue = [72362, 55912, 84511, 23961, 33922]
labels = ['Vegetables', 'Meat', 'Snacks', 'Drinks', 'Fruits']
explode = [0,0.1,0,0,0]

plt.pie(revenue, labels=labels, wedgeprops={'edgecolor':'black'},explode=explode, startangle=180)
plt.title('Food Sales')
```
